```python
#Lower casing and removing punctuations

df['Text'] = df['Text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
df['Text'] = df['Text'].str.replace('[^\w\s]', "")
df.Text.head()
```

```python
#Removal of stop words

stop = stopwords.words('english')
df['Text'] = df['Text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df.Text.head()
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
#Lemmatization (Sorting and grouping similar words)

df['Text'] = df['Text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df.Text.head
```

```python
#Exploratory data analysis
reviews = df

#Dropping null values
reviews.dropna(inplace=True)
```

```python
score_1 = reviews[reviews['Score'] == 1]
score_2 = reviews[reviews['Score'] == 2]
score_3 = reviews[reviews['Score'] == 3]
score_4 = reviews[reviews['Score'] == 4]
score_5 = reviews[reviews['Score'] == 5]
```

```python
review_sample = pd.concat([score_1, score_2, score_3, score_4, score_5], axis = 0)
review_sample.reset_index(drop = True, inplace=True)
```

```python
reviews_str = review_sample.Summary.str.cat()
wordcloud = WordCloud(background_color = 'white').generate(reviews_str)
plt.figure(figsize = (10,10))
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
negative_review = review_sample[review_sample['Score'].isin([1,2])]
positive_review = review_sample[review_sample['Score'].isin([4,5])]

#Transfor to single string
negative_review_str = negative_review.Summary.str.cat()
positive_review_str = positive_review.Summary.str.cat()
```

```python
#Create WordCloud

wordcloud_negative = WordCloud(background_color = 'white').generate(negative_review_str)
wordcloud_positive = WordCloud(background_color = 'black').generate(positive_review_str)

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(211)
ax1.imshow(wordcloud_negative, interpolation = 'bilinear')
ax1.axis("off")
ax1.set_title('Reviews with negative scores', fontsize = 20)
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(212)
ax1.imshow(wordcloud_positive, interpolation = 'bilinear')
ax1.axis("off")
ax1.set_title('Reviews with positive scores', fontsize = 20)
```

```python
import re
import os
import sys
import ast
plt.style.use('fivethirtyeight')

#Function for getting the sentiment
cp = sns.color_palette()

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
#Generating sentiment for all the sentences present in the dataset
emptyline=[]
for row in df['Text']:
    vs =  analyzer.polarity_scores(row)
    emptyline.append(vs)

#Creating new dataframe with sentiments

df_sentiments = pd.DataFrame(emptyline)
df_sentiments.head()
```

```python
#Merging the sentiments back to reviews dataframe
df_c = pd.concat([df.reset_index(drop = True), df_sentiments], axis = 1)
df_c.head(3)
```

```python
#Convert scores into positive and negative sentiments using some threshold
df_c['Sentiment'] = np.where(df_c['compound'] >= 0, 'Positive', 'Negative')
df_c.head(5)
```

```python
#Final Result
result = df_c['Sentiment'].value_counts()
result.plot(kind = 'bar', rot = 0, color=['plum', 'cyan']);
```
